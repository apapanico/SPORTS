```python
games = Table().read_table('ncaa16games.csv')
games
```

```python
dates = make_array()
for i in np.arange(len(games[2])):
    dates = np.append(dates, games[2][i].split(' '))
zero = np.arange(0, 2619, 3)
one = np.arange(1, 2619, 3)
two = np.arange(2, 2619, 3)
month = dates[zero]
day = dates[one]
years = dates[two]
```

```python
months = make_array()
for i in month:
    if i == "Aug":
        months = np.append(months, "08")
    elif i == 'Sep':
        months = np.append(months, "09")
    elif i == 'Oct':
        months = np.append(months, "10")
    elif i == 'Nov':
        months = np.append(months, "11")
    elif i == 'Dec':
        months = np.append(months, "12")
    elif i == 'Jan':
        months = np.append(months, "01")
    else:
        months = np.append(months, "NA")
```

```python
days = make_array()
for i in day:
    if i == "1":
        days = np.append(days, "01")
    elif i == '2':
        days = np.append(days, "02")
    elif i == '3':
        days = np.append(days, "03")
    elif i == '4':
        days = np.append(days, "04")
    elif i == '5':
        days = np.append(days, "05")
    elif i == '6':
        days = np.append(days, "06")
    elif i == '7':
        days = np.append(days, "07")
    elif i == '8':
        days = np.append(days, "08")
    elif i == '9':
        days = np.append(days, "09")
    else:
        days = np.append(days, i)
```

```python
games = games.with_columns('m', months, 'd', days, 'y', years)
#games
```

```python
team1 = make_array()
team2 = make_array()
team1f = make_array()
team2f = make_array()
for i in np.arange(len(games[5])):
    if "@" in games[7][i]:
        if (games[5][i][0] == '(') and (games[8][i][0] == '('):
            team1 = np.append(team1, games[5][i].split(')\xa0')[1])
            team2 = np.append(team2, games[8][i].split(')\xa0')[1])
        elif (games[5][i][0] == '(') and (games[8][i][0] != '('):
            team1 = np.append(team1, games[5][i].split(')\xa0')[1])
            team2 = np.append(team2, games[8][i])
        elif (games[5][i][0] != '(') and (games[8][i][0] == '('):
            team1 = np.append(team1, games[5][i])
            team2 = np.append(team2, games[8][i].split(')\xa0')[1])
        else:
            team1 = np.append(team1, games[5][i])
            team2 = np.append(team2, games[8][i])
    else:
        if (games[8][i][0] == '(') and (games[5][i][0] == '('):
            team1 = np.append(team1, games[8][i].split(')\xa0')[1])
            team2 = np.append(team2, games[5][i].split(')\xa0')[1])
        elif (games[8][i][0] == '(') and (games[5][i][0] != '('):
            team1 = np.append(team1, games[8][i].split(')\xa0')[1])
            team2 = np.append(team2, games[5][i])
        elif (games[8][i][0] != '(') and (games[5][i][0] == '('):
            team1 = np.append(team1, games[8][i])
            team2 = np.append(team2, games[5][i].split(')\xa0')[1])
        else:
            team1 = np.append(team1, games[8][i])
            team2 = np.append(team2, games[5][i])
for i in np.arange(len(games[5])):
    if "State" in team1[i]:
        team1f = np.append(team1f, team1[i].replace("State", "st").replace(' ', '-').lower())
    else:
        team1f = np.append(team1f, team1[i].replace(' ', '-').replace('&', '').lower())
    if "State" in team2[i]:
        team2f = np.append(team2f, team2[i].replace("State", "st").replace(' ', '-').lower())
    else:
        team2f = np.append(team2f, team2[i].replace(' ', '-').replace('&', '').lower())
                       
games = games.with_columns('team1', team1f, 'team2', team2f)
#games
```

```python
#!pip install requests
import requests
import json

f = open("links2016.txt", "r")
urls = []
# this assumes that each url is on a new line in the url-file
url = f.readline()
while (url):
    # if there's a comma at the end of the url, do `urls.append(url[:-2])`. You should also get rid of quotes and stuff
    urls.append(url[:-1])
    url = f.readline()
f.close()

for url in urls:
    r = requests.get(url)
    if (r.status_code != 200):
        # it couldn't get the web page
        print ("failed to get" + url)
        print (r.status_code)
        continue
    d = json.loads(str(r.text))
    year = url.split("/")[6]
    filename = "playbyplay/"+d['meta']['teams'][0]['seoName'] + "-vs-" + d['meta']['teams'][1]['seoName'] + "-" + year
    with open(filename+'.json', 'w') as outfile:
        json.dump(d, outfile)
    print("wrote to "+filename)
```
